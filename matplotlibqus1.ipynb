```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
delivery=pd.read_csv("deliveries.csv")

delivery
```

```python
matches=pd.read_csv("matches.csv")
matches
```

```python
new=delivery.merge(matches,left_on='match_id',right_on='id')
```

```python
new
```

```python
def total(batsman_name):
    runs=new.groupby(['season','batsman'])['batsman_runs'].sum().reset_index()
    mask=runs['batsman']==batsman_name
    s=runs[mask]
    s.plot(kind='line',x='season',y='batsman_runs',color='red',marker='*')
    return plt.show()

    
    
   
    

    
    


    
```

```python
total('V Kohli')
```
