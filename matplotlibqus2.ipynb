```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
delivery=pd.read_csv("deliveries.csv")

delivery
```

```python
def scoring_area(batsman_name):
    runs=delivery.groupby('batsman')
    s=runs.get_group(batsman_name)
    analysis=s['batsman_runs'].value_counts()
    return analysis.plot(kind='pie',autopct="%0.1f%%")
```

```python
scoring_area('CH Gayle')
```
